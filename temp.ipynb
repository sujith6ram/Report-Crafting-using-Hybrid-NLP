```python
from happytransformer import  HappyTextToText
happy_tt = HappyTextToText("T5", "vennify/t5-base-grammar-correction")

from happytransformer import TTSettings
beam_settings =  TTSettings(num_beams=5, min_length=1, max_length=100)

def gram(paragraph):
  # Split the paragraph into sentences
  sentences = paragraph.split('.')
  c=""
  # Print the sentences
  for i, sentence in enumerate(sentences, 1):
    # print(f"Sentence {i}: {sentence.strip()}")
    output_text_1 = happy_tt.generate_text(sentence, args=beam_settings)
    # print(f"Sentence {i}: {output_text_1.text.strip()}")
    c+=" "+output_text_1.text.strip()
  return c
```

```python
from docx2pdf import convert
from PyPDF2 import PdfMerger

def convert_and_merge(docx_files, output_pdf):
    # Convert Word documents to PDF
    pdf_files = []
    for docx_file in docx_files:
        pdf_file = docx_file.replace('.docx', '.pdf')
        convert(docx_file, pdf_file)
        pdf_files.append(pdf_file)

    # Merge PDF files
    merger = PdfMerger()
    for pdf_file in pdf_files:
        merger.append(pdf_file)
    merger.write(output_pdf)
    merger.close()

# Usage
docx_files = ['ex.docx','ex1.docx', 'ex2.docx',]
output_pdf = 'REPORT CRAFTING USING HYBRID NLP FINAL1.pdf'
convert_and_merge(docx_files, output_pdf)
```

```python
from docx2pdf import convert
from PyPDF2 import PdfMerger

def convert_and_merge(docx_files, output_pdf):
    # Convert Word documents to PDF
    pdf_files = []
    for docx_file in docx_files:
        pdf_file = docx_file.replace('.docx', '.pdf')
        convert(docx_file, pdf_file)
        pdf_files.append(pdf_file)

    # Merge PDF files
    merger = PdfMerger()
    for pdf_file in pdf_files:
        merger.append(pdf_file)
    merger.write(output_pdf)
    merger.close()

# Usage
docx_files = ['exhw.docx','exhw1.docx', 'exhw2.docx',]
output_pdf = 'WATER QUALITY DETECTION AND MANAGEMENT SYSTEM.pdf'
convert_and_merge(docx_files, output_pdf)
```
